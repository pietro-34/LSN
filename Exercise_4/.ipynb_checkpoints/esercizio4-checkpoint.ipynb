```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("gas/OUTPUT/pofv.dat")
```

```python


# ultimo blocco
blk_finale = np.max(data[:,0])
data_finale = data[data[:,0] == blk_finale]

v_bin_center = data_finale[:,1]
pofv_prog = data_finale[:,3]
errori=data_finale[:,4]
plt.bar(v_bin_center, pofv_prog, width=v_bin_center[1] - v_bin_center[0], align='center', alpha=0.7, label='POFV media progressiva')
plt.errorbar(v_bin_center, pofv_prog, yerr=errori, fmt='none', ecolor='red', capsize=3)


plt.xlabel("v")
plt.ylabel("P(v)")
plt.title(f"POFV - Istogramma (Blocco {int(blk_finale)})")
plt.grid(True)
plt.legend()
plt.show()
```

```python

import imageio
import os

# Parametri
#input_file = "dirac/OUTPUT/pofv.dat" già letto nella cella precedente
output_gif = "pofv_evolution.gif"
n_blocchi = 20
img_dir = "frames"

# Crea cartella temporanea per i frame
os.makedirs(img_dir, exist_ok=True)

# Carica tutti i dati
#data = np.loadtxt(input_file) già letto nella cerlla precedente e salvato in data2

# Lista dei file immagine
filenames = []

# Crea un frame per ogni blocco
for blk in range(1, n_blocchi + 1):
    data_blk = data2[data2[:, 0] == blk]

    v_bin_center = data_blk[:, 1]
    pofv_prog = data_blk[:, 3]
    errori = data_blk[:, 4]

    plt.figure(figsize=(8, 5))
    plt.bar(v_bin_center, pofv_prog, width=v_bin_center[1]-v_bin_center[0],
            align='center', alpha=0.7, color='skyblue', label='POFV media progressiva')
    plt.errorbar(v_bin_center, pofv_prog, yerr=errori,
                 fmt='none', ecolor='red', capsize=3)

    plt.xlabel("v")
    plt.ylabel("P(v)")
    plt.title(f"Distribuzione POFV - Blocco {blk}")
    plt.grid(True)
    plt.ylim(0, max(pofv_prog)*1.3)
    plt.legend()

    # Salva il frame
    filename = f"{img_dir}/frame_{blk:03d}.png"
    plt.savefig(filename)
    filenames.append(filename)
    plt.close()

# Crea la GIF
with imageio.get_writer(output_gif, mode='I', duration=1.2) as writer:
    for filename in filenames:
        image = imageio.imread(filename)
        writer.append_data(image)

# Pulizia (opzionale)
# for filename in filenames:
#     os.remove(filename)
# os.rmdir(img_dir)

print(f"GIF creata: {output_gif}")
```

```python
N = 20  # numero di blocchi
x = np.arange(N)

# Carica dati
press, error_press = np.loadtxt("gas/OUTPUT/pressure.dat", usecols=(2,3), unpack=True)
ekin, error_ekin   = np.loadtxt("gas/OUTPUT/kinetic_energy.dat", usecols=(2,3), unpack=True)
epot, error_epot   = np.loadtxt("gas/OUTPUT/potential_energy.dat", usecols=(2,3), unpack=True)
etot, error_etot   = np.loadtxt("gas/OUTPUT/total_energy.dat", usecols=(2,3), unpack=True)
temp, error_temp   = np.loadtxt("gas/OUTPUT/temperature.dat", usecols=(2,3), unpack=True)

# Configura plot principale e axes
fig, axes = plt.subplots(3, 2, figsize=(20, 18))
fig.suptitle("Simulation", fontsize=22)

# Pressione
axes[0, 0].errorbar(x, press, yerr=error_press, label="simulation")
axes[0, 0].set_title("Pression")
axes[0, 0].set_xlabel("Blocks")
axes[0, 0].set_ylabel("P*")
axes[0, 0].hlines(press[-1], 0, N, color="yellow", label="last block value")
axes[0, 0].grid(True)

# Energia cinetica
axes[0, 1].errorbar(x, ekin, yerr=error_ekin)
axes[0, 1].set_title("Kinetic Energy")
axes[0, 1].set_xlabel("Blocks")
axes[0, 1].set_ylabel("K*")
axes[0, 1].hlines(ekin[-1], 0, N, color="yellow")
axes[0, 1].grid(True)

# Energia potenziale
axes[1, 0].errorbar(x, epot, yerr=error_epot)
axes[1, 0].set_title("Potential Energy")
axes[1, 0].set_xlabel("Blocks")
axes[1, 0].set_ylabel("V*")
axes[1, 0].hlines(epot[-1], 0, N, color="yellow")
axes[1, 0].grid(True)

# Energia totale
axes[1, 1].errorbar(x, etot, yerr=error_etot)
axes[1, 1].set_title("Total Energy")
axes[1, 1].set_xlabel("Blocks")
axes[1, 1].set_ylabel("E*")
axes[1, 1].hlines(etot[-1], 0, N, color="yellow")
axes[1, 1].grid(True)

# Temperatura
axes[2, 0].errorbar(x, temp, yerr=error_temp)
axes[2, 0].set_title("Temperature")
axes[2, 0].set_xlabel("Blocks")
axes[2, 0].set_ylabel("T*")
axes[2, 0].hlines(temp[-1], 0, N, color="yellow")
axes[2, 0].hlines(1.2, 0, N, color="red", label="desired temperature")
axes[2, 0].grid(True)

# Ultima subplot vuota per legenda globale
axes[2, 1].axis("off")

# Legenda globale
handles, labels = axes[0, 0].get_legend_handles_labels()
fig.legend(handles, labels, fontsize=15, loc='lower right')

plt.tight_layout(rect=[0, 0, 1, 0.96])  # lascia spazio per titolo globale
plt.show()
```
