```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
X, Y, Z =np.loadtxt("prob_1s.dat", usecols=(0,1,2), unpack=True, skiprows=1)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(X, Y, Z, c=Z, marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(10, 30)

plt.show()
```
