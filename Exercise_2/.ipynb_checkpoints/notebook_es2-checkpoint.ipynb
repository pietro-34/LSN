```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
M=100000
N=100
L= M/N
```

```python
medie = np.loadtxt('output_medie.txt')
errori=np.loadtxt('output_errori.txt')
M=100000
N=100
L= M/N
x = np.arange(N)  
x= x*L
#plt.plot(x,medie-0.5)
plt.errorbar(x,medie-1,yerr=errori, ecolor='red')
plt.xlabel('#throws')
plt.ylabel(r'$ < \int_{0}^{1} \frac{\pi}{2} \cos(\frac{\pi}{2} x) > -1 $')
plt.grid(True)
plt.show()
```

```python
medie_sampling = np.loadtxt('output_medie_sampling.txt')
errori_sampling=np.loadtxt('output_errori_sampling.txt')

plt.errorbar(x,medie_sampling-1,yerr=errori_sampling, ecolor='red')
plt.xlabel('#throws')
plt.ylabel(r'$ < I >_{d} $')
plt.grid(True)
plt.show()
```
