### Equilibrazione simulazioni dinamica molecolare

#### Impostazione della temperatura iniziale

Il sistema, costituito da 108 atomi, è inizializzato in una configurazione spaziale di tipo reticolo FCC, con velocità distribuite secondo una legge di Maxwell-Boltzmann.

All'inizio della simulazione, gli atomi occupano posizioni corrispondenti al minimo del potenziale di Lennard-Jones (LJ), pertanto l'energia potenziale totale $U$ si trova prossima al suo valore minimo.  
In questa configurazione, l'energia potenziale costituisce una frazione trascurabile dell'energia totale, che, essendo la simulazione condotta nell'ensemble microcanonico (NVE), si mantiene costante.

L'attribuzione di velocità iniziali genera un processo di rilassamento dinamico: gli atomi si allontanano dalle posizioni di equilibrio, provocando un incremento dell'energia potenziale a spese dell'energia cinetica. Poiché la temperatura è proporzionale all'energia cinetica media, si osserva una diminuzione della temperatura durante questa fase.

Per consentire al sistema di raggiungere la temperatura desiderata di $T^* = 1.1$ al termine del rilassamento, la temperatura iniziale è stata impostata a un valore maggiore. In particolare, si è riscontrato che una temperatura iniziale di $T^* = 2.0$ consente di ottenere un buon bilanciamento.


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
istant_temp_md = np.loadtxt("moleculardynamics/OUTPUT/istant_temperature.dat")
#istant_magnet_random = np.loadtxt("equilibration/OUTPUT/istant_magnetization_random.dat")
```

```python
nsteps= 5000
x=np.arange(nsteps)


plt.figure(figsize=(12, 6))
# Plot principale
plt.plot(x, istant_temp_md, label="Istant Temperature")

# Retta orizzontale a y=1.1
plt.axhline(y=1.1, color="red", linestyle="--", linewidth=1, label=r"$T^*=1.1$")

# Etichette e titolo
plt.xlabel("# steps", fontsize=14)
plt.ylabel(r"$T^*$", fontsize=14)
plt.title("Istant Temperature", fontsize=16)

plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

#### Evoluzione della temperatura istantanea

Il grafico mostra una brusca diminuzione della temperatura nei primi $100$ step della simulazione, seguita da un regime oscillatorio attorno alla temperatura desiderata $T^* = 1.1$.

Le oscillazioni in ampiezza si riducono leggermente dopo circa $10^3$ step, suggerendo che il sistema sta raggiungendo un regime di equilibrio termico.

Sulla base di queste osservazioni, ho deciso di scartare i primi $10^3$ step della simulazione nel calcolo dei valori medi.  
In pratica, il sistema è stato simulato per $10^3$ step iniziali, dopo i quali è stato eseguito un **RESTART** da tale punto, proseguendo l’evoluzione del sistema già in condizioni più prossime all’equilibrio.
