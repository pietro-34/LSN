```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
import pandas as pd
```

```python
M=100000
N=100
L= M/N
x = np.arange(N)  
x= x*L
```

```python
data_call=pd.read_csv("output_call.txt", sep="\t")
data_call=data_call.drop('Unnamed: 4',axis=1)
data_call
```

```python
data_put=pd.read_csv("output_put.txt", sep="\t")
data_put=data_put.drop('Unnamed: 4',axis=1)
data_put
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))  # crea 1 riga e 2 colonne

# Plot diretto sulla prima colonna
axes[0].errorbar(x, data_call['CALL_dir'], yerr=data_call['Err_CALL_dir'], ecolor='red')
axes[0].set_title('Direct sampling')
axes[0].set_xlabel('#GBM')
axes[0].set_ylabel(r'$<Call>$')
axes[0].grid(True)

# Plot indiretto sulla seconda colonna
axes[1].errorbar(x, data_call['CALL_ind'], yerr=data_call['Err_CALL_ind'], ecolor='red')
axes[1].set_title('Indirect sampling')
axes[1].set_xlabel('#GBM')
axes[1].set_ylabel(r'$<Call>$')
axes[1].grid(True)

plt.tight_layout()  # migliora l'aspetto evitando sovrapposizioni
plt.show()
```

```python

fig, axes = plt.subplots(1, 2, figsize=(12, 5))  # crea 1 riga e 2 colonne

# Plot diretto sulla prima colonna
axes[0].errorbar(x, data_put['PUT_dir'], yerr=data_put['Err_PUT_dir'], ecolor='red')
axes[0].set_title('Direct sampling')
axes[0].set_xlabel('#GBM')
axes[0].set_ylabel(r'$<Put>$')
axes[0].grid(True)

# Plot indiretto sulla seconda colonna
axes[1].errorbar(x, data_put['PUT_ind'], yerr=data_put['Err_PUT_ind'], ecolor='red')
axes[1].set_title('Indirect sampling')
axes[1].set_xlabel('#GBM')
axes[1].set_ylabel(r'$<Put>$')
axes[1].grid(True)

plt.tight_layout()  # migliora l'aspetto evitando sovrapposizioni
plt.show()
```
