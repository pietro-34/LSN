SIMULAZIONE NVT di un modello di ising 1D; N=50 T=2.

Campionamento con metropolis.
L'equilibrazione del sistema avviene in pochi passi anche nel caso in cui il sistema è una configurazione di spin tutti up, come mostra il primo grafico. Questom perchè per ogni step della simulazion e montecarlo ha già fatto 50 passi.

Prendendo i blocchi sufficientemente lunghi #steps> 100  e con l'evidenza della veloce convergenza all'equilibrio, mostrata nei precedenti grafici,  non è necessario scartare alcun transiente iniziale, per il calcolo delle medie progressive

Con queste considerazioni ho scelto di simulare il sistema in 20 blocchi da 10000 steps ciascuno.

```python
#magnetizzazione teorica:
points=100
T = np.linspace(0.2,3.0,num=points)
beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th

h=0.02 #external field
b = 1/T

 #energy per particle
e = -J*( th + ch*thN )/( 1 + thN ) #energy per particle




l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns
#magnetization
M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)

#susceptibility
X = beta*np.exp(2*beta*J)*(1-thN)/(1+thN)

#specific_heat
heat=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)
```

```python
Temp, energy, err_energy =np.loadtxt("metropolis/output_metropolis/energy.dat", usecols=(0,1,2), unpack=True)
```

```python
#plot energy metropolis vs teorica
plt.figure(figsize=(12,8))   
#plot
plt.errorbar(Temp, energy, yerr=err_energy,
             fmt='o',                         # solo simboli, no linea
             ecolor='red',                    # colore barre errore
             capsize=5,
             markerfacecolor='green',
             markeredgecolor='green',
             label='simulation')

# Curva teorica: linea blu continua
plt.plot(T, e, color='blue', label='teorica')
plt.title("Energy per particle ", fontsize=14)
plt.legend()
plt.xlabel("Temperature", fontsize=14)
plt.ylabel(r"$ \frac{< H >}{N} $", fontsize=14)
#plt.hlines(press[-1], 0, N, color="yellow", label="last block value")
plt.grid(True)
```
