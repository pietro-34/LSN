```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
medie = np.loadtxt('output_medie.txt')
errori=np.loadtxt('output_errori.txt')
N=1000
x = np.arange(N)  
plt.plot(x,medie-0.5)
#plt.errorbar(x*10,0.5-medie,yerr=errori)
```

```python

# 2. Crea l'asse x (da 1 a 1000 per il numero di blocchi)
# Se hai 1000 blocchi, il numero di numeri generati è 10000, quindi x va da 0 a 10000
blocchi = np.linspace(1, 1000, len(medie))

# 3. Crea il grafico con Matplotlib
plt.errorbar(blocchi, medie-0.5, yerr=errori, fmt='o', ecolor='r', capsize=3, linestyle=' ', color='b', label='Media su blocchi')

# 4. Etichette e titolo
plt.xlabel('Numero di blocchi')
plt.ylabel('Media sul blocco')
plt.title('Media su blocchi con barre di errore')
plt.legend()

# 5. Mostra il grafico
plt.show()
```
