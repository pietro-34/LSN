```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
M=100000
N=100
L= M/N
```

```python
medie = np.loadtxt('output_medie.txt')
errori=np.loadtxt('output_errori.txt')
M=100000
N=100
L= M/N
x = np.arange(N)  
x= x*L
#plt.plot(x,medie-0.5)
plt.errorbar(x,medie-0.5,yerr=errori, ecolor='red')
plt.xlabel('#throws')
plt.ylabel('<r>-1/2')
plt.grid(True)
plt.show()
```

```python
medie_var = np.loadtxt('output_var_medie.txt')
errori_var=np.loadtxt('output_errori_var.txt')

l=1/12
plt.errorbar(x,medie_var-l,yerr=errori_var, ecolor='red')
plt.xlabel('#throws')
plt.ylabel('<(r-0.5)^2>-1/12')
plt.grid(True)
plt.show()
```
